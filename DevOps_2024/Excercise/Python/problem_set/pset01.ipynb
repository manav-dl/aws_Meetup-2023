## Problems

Problem 1:
You are given a list of mixed data types: [1, "apple", 3.14, True]. Additionally, you have a string "banana". Write a Python program that adds the string to the end of the list. After adding, output the updated list to verify the addition.

Problem 2:
Consider two lists: ["cat", "dog", "fish"] and ["bird", "hamster", "rabbit"]. Your task is to write a Python program that concatenates these two lists into a single, unified list. Once concatenated, output the resulting list to ensure all elements from both original lists are included.

Problem 3:
Given a list containing a variety of data types: [42, "hello", 3.5, "world", False, 99], write a Python program that creates a new list which consists of only the first four elements of the original list. Output the new list to verify it contains the correct subset of elements.

Problem 4:
You have a list: ["red", "green", "blue", "yellow", "green", "purple"] and you need to remove the string "green". Write a Python program that removes the first occurrence of this string from the list. Output the updated list to confirm the removal was successful and only the first instance of "green" was removed.

Problem 5:
You are provided with a list of mixed elements: ["Python", 100, "Java", 200, "C++", 300]. Write a Python program that accesses the third element of the list and prints it. Additionally, access the last element of the list and print it. Output both elements to verify correct access.
